```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", 34)

nutritionobesitydf = pd.read_csv('/content/drive/MyDrive/Citadel Datathon 2024 team13/Nutrition_Physical_Activity_and_Obesity_Data.csv')

nutritionobesitydf.head()
```

```python
for column in nutritionobesitydf.columns:
  if pd.api.types.is_numeric_dtype(nutritionobesitydf[column]):
    nutritionobesitydf[column] = nutritionobesitydf[column].interpolate()
nutritionobesitydf.head()
```

```python
obesity = nutritionobesitydf[nutritionobesitydf['Question'] == 'Percent of adults aged 18 years and older who have obesity']
```

```python
obesity[(obesity["YearStart"] == 2018) & (obesity["LocationAbbr"] == "TX")]
```

```python
obesity_filtered = obesity[obesity["YearStart"] == 2018][["YearStart", "LocationAbbr","Data_Value", "Sample_Size"]]
```

```python
obesity_filtered.head()
```

```python
obesity_filtered['WeightedValue'] = obesity_filtered['Data_Value'] * obesity_filtered['Sample_Size']

grouped = obesity_filtered.groupby('LocationAbbr').agg({'WeightedValue': 'sum', 'Sample_Size': 'sum'})

grouped['WeightedAverage'] = grouped['WeightedValue'] / grouped['Sample_Size']

print(grouped)
```

```python
obesity_filtered = grouped.reset_index()
```

```python
obesity_filtered = obesity_filtered[["LocationAbbr", "WeightedAverage"]]
```

```python
law_df = pd.read_excel("weighted_food_safety_laws_by_state_and_year.xlsx")
```

```python
law_df = law_df.T
law_df.columns = law_df.iloc[0]
law_df_filtered = law_df[[2018.0]]
law_df_filtered = law_df_filtered[1:]
law_df_filtered = law_df_filtered.reset_index()

law_df_filtered.columns = ["LocationAbbr","Percetage"]
law_df_filtered
```

```python
merged_df = pd.merge(law_df_filtered, obesity_filtered, on = ["LocationAbbr"], how = "inner")
```

```python
valid_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
    'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
    'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

filtered_df = merged_df[merged_df['LocationAbbr'].isin(valid_states)]
```

```python
from google.colab import files
csv_path = 'Cleaned_state_percentages.csv'
filtered_df.to_csv(csv_path, index=False)
print(f"DataFrame is saved to {csv_path}")

files.download(csv_path)
```
